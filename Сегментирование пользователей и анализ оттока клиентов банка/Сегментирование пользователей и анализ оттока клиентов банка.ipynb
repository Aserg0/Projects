````python
def get_df_name(df):            #функция получения имени проверяемого датафрейма
    name =[x for x in globals() if globals()[x] is df][0]
    return name

#Функция проверки датафрейма на дубликаты, пропуски, вывод информации 
#о датафрейме, описат. статистики
def checking_data(df): 
    
    print('Данные по датафрейму', get_df_name(df), ':')
    print()
    display(df.head())
    #Проверим наличие дубликатов строк
    print('Дубликатов строк в датафрейме:', df.duplicated().sum()) 
    print('')
    print('Количество пропусков в данных в столбцах:')
    empty_cells = df.isna().agg(['sum', 'mean']).T.rename(columns={'sum': 'total_blanks', 'mean': 'share_blanks, %'})
    empty_cells['total_blanks'] = empty_cells['total_blanks'].astype(int)
    empty_cells['share_blanks, %'] = round(empty_cells['share_blanks, %']*100, 2)
    display(empty_cells)
    print('')
    print('_'*25, 'dataframe info: ', '_'*25)
    # изучим информацию о типах данных в столбцах и размере датафрейма, 
    #возможных пропусках в данных 
    df.info() 
    print('')
    #попытаемся понять, есть ли аномальные значения
    print('Описательная статистика датафрейма:')
    display(df.describe(include = "all").round(2)); 
````

````python
#проверяем датафрейм logs на дубликаты, пропуски, вывод информации 
#о датафрейме,описат. статистики
print(checking_data(bank_clients));
````

### Обработка пропусков в колонке age

````python
#запишем userid строк с пропусками в колонке age в отдельный список  |
empty_age=bank_clients[bank_clients['age'].isna()]['userid'].reset_index(drop=True)
````

````python
#запишем userid строк с дубликатами userid в отдельный список. 
dupl_userid=bank_clients[bank_clients.duplicated(subset=['userid'])]['userid'].reset_index(drop=True)
````

````python
#выведем строки с пропущенным значением в колонке age, но имеющие дубликат userid
display(bank_clients[(bank_clients['userid'].isin(empty_age))&(bank_clients['userid'].isin(dupl_userid))]\
       .sort_values(by='userid', ascending=True))
````

Среди клиентов с пропуском в колонке age только у клинета с userid 141945 (с индексом 9457) полное совпадение профиля с данными в строке-дубликате по userid, если рассматривать колонки score, gender, products, credit_cards, last_activity, est_salary. Таким образом можно говорить, что это один и тот же клиент, но находящийся в двух филиалах - Рыбинске и Ярославле, и значение для пропуска в колонке age взять из строки-дубликата.    
Также для целей данного анализа будем считать, что клиента с userid 141945 можем оставить только в Ярославле, а строку, связанную с Рыбинском (индекс 4343), удалим, т.к. это полный дубликат.  
По остальным клиентам, есть похожие строки-дубликаты, но они не так явно совпадают, и это, скорее всего, разные клиенты. Поэтому взять данные для заполнения колонки age мы оттуда не можем.

````python
#указываем значение из строки-дубликата для пропуска в колонке age по
#клиенту с userid 141945 (с индексом 9457) 
bank_clients.loc[9457, 'age']=33
````

````python
def diagramms(df, hue=None): #Функция построения диаграмм распределения параметров, в т.ч. с учетом параметра hue.

    #распределения количественных переменных с учетом количества продуктов
    for column in ['score', 'age', 'est_salary', 'balance']:
        plt.figure(figsize= (10, 4))
        if column == 'est_salary':
            plt.xlim([0, 400000])
            sns.histplot(data=df, x=column, stat='density', hue=hue, kde=True, common_norm=False)
        elif column == 'balance':
            plt.xlim([0, 5000000])
            sns.histplot(data=df[df['balance'].notnull()], x=column, stat='density', hue=hue, \
                     kde=True, common_norm=False)
        else:
            sns.histplot(data=df, x=column, binwidth=1, stat='density', hue=hue, kde=True, common_norm=False)
        plt.ylabel('Плотность распределения')
        plt.xlabel('Значения параметра ' + column + ' с учетом количества продуктов')
        plt.title('Распределение параметра '+ column + ' с учетом параметра '  + hue) 
        plt.show()
    
    #Построим диаграммы для категориальных признаков с учетом количества продуктов
    for column in ['city', 'gender', 'equity', 'products', 'credit_card', 'last_activity', 'churn']:
        plt.figure(figsize=(6,3))
        sns.countplot(data=df, x=column, hue=hue)
        plt.xlabel('Значения параметра ' + column)
        plt.ylabel('Количество клиентов')
        plt.title('Распределение клиентов по признаку ' + column + ' с учетом параметра ' + hue)
        plt.show()
    
    
    print();
````

## Статистическая проверка гипотез

###  Гипотеза равенства возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним продуктом

Для проверки гипотезы сформируем две выборки данных в зависмости от количества продуктов: с 2 продуктами и 1 продуктом.
В этих выборках будем сравнивать значения возраста клиентов.

````python
bank_clients_new1 = bank_clients_new.query('products == 1')['age']
bank_clients_new2 = bank_clients_new.query('products == 2')['age']
````

По этим данным гипотезу можно проверить: 
- как методом для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам 
t-тест (st.ttest_ind (array1, array2, equal_var=True)), 
- так и непараметрическим тестом Уилкоксона-Манна-Уитни по критерию Манна-Уитни (st.mannwhitneyu(dataset_1, dataset_2, True, 'two-sided')), который проверяет однородность распределения параметров.

Для первого метода проверки, важны три аспекта, которые надо проверить, чтобы понять, можно ли применять t-тест:
1. Генеральные совокупности не должны зависеть друг от друга. 
2. Выборочные средние должны быть нормально распределены.
3. Дисперсии рассматриваемых генеральных совокупностей должны быть равны.

````python
#проверка данных на нормальность распределения

for val in [bank_clients_new1, bank_clients_new2]:
    alpha = 0.05  # критический уровень статистической значимости

    results = st.shapiro(val)
    p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

    print('p-значение: ', p_value)

    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: распределение не нормально')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
````

Мы имеем два независимых набора данных, которые не распределены нормально, поэтому для t-теста применяем значение equal_var=False

````python
# проверим значения дисперсий в данных
print(np.var(bank_clients_new1))
print(np.var(bank_clients_new2))
````

Дисперсии в данных похожи, но не равны.

**Сформулируем основную гипотезу и ее альтернативу:**  
- Нулевая гипотеза H_0: Возраст клиентов, которые пользуются двумя продуктами банка, и тех, которые пользуются одним продуктом, не отличается.  
- Альтернативная гипотеза H_1: Возраст клиентов, которые пользуются двумя продуктами банка, и тех, которые пользуются одним продуктом, отличается.  

Критический уровень статистической значимости примем в размере alpha = 0.05

Сначала проверим отличается ли средний возраст клиентов с 1 и 2 продуктам с помощью t-теста.

````python
#проверим гипотезу о равенстве среднего возраста в группах двусторонним t-тестом Стьюдента

alpha = 0.05
results = st.ttest_ind(bank_clients_new1, bank_clients_new2, equal_var=False)

print('p-значение:', "{0:.4f}".format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
````

Теперь проверим изменился ли возраст клиентов с 2 продуктами по сравнению с клиентами с 1 продуктом.

````python
#Посчитаем статистическую значимость различия в возрасте между группами.
#Применим критерий Манна-Уитни и отформатируем p-value, округлив его до 4 знаков после запятой.

alpha = 0.05 # критический уровень статистической значимости
results = st.mannwhitneyu(bank_clients_new1, bank_clients_new2)
print('p-значение: ',"{0:.4f}".format(results[1]))

if results[1] < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
````

Проверка гипотезы t-тестом показала, что средний возраст клиентов с 1 продуктом и двумя продуктами отличается. В тоже время проверка гипотезы криетрием Манна-Уитни отличий в возрасте сравниваемых групп данных не выявила.   
Учитывая, что критерий Манна-Уитни менее чувствителен к выбросам в данных, а также то, что наши данные не полностью отвечают требованиям проведения t-теста (распределены данные не нормально), в данном случае критерий Манна-Уитни больше подходит для проверки вышеуказанной гипотезы.   
**Таким образом, согласно непараметрического теста Уилкоксона-Манна-Уитни нет статистически значимой разницы в возрасте клиентов, которые пользуются двумя продуктами банка, и клиентов, которые пользуются одним продуктом.** 

###  Гипотеза равенства оценочной заработной платы у клиентов с 2 продуктами и клиентов с 1 продуктом.

````python
#проверка данных на нормальность распределения

for val in [bank_clients_news1, bank_clients_news2]:
    alpha = 0.05  # критический уровень статистической значимости

    results = st.shapiro(val)
    p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

    print('p-значение: ', p_value)

    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: распределение не нормально')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
````
